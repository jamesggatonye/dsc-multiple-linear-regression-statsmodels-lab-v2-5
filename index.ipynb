## The Ames Housing Dataset

The [Ames Housing dataset](http://jse.amstat.org/v19n3/decock.pdf) is a newer (2011) replacement for the classic Boston Housing dataset. Each record represents a residential property sale in Ames, Iowa. It contains many different potential predictors and the target variable is `SalePrice`.

````python
import pandas as pd
ames = pd.read_csv("ames.csv", index_col=0)
ames
````

````python
ames.describe()
````

We will focus specifically on a subset of the overall dataset. These features are:

```
LotArea: Lot size in square feet

1stFlrSF: First Floor square feet

GrLivArea: Above grade (ground) living area square feet
```

````python
ames_subset = ames[['LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']].copy()
ames_subset
````

## Step 1: Visualize Relationships Between Features and Target

For each feature in the subset, create a scatter plot that shows the feature on the x-axis and `SalePrice` on the y-axis.

````python
# Your code here - import relevant library, create scatter plots
import matplotlib.pyplot as plt
# plotting area
plt.figure(figsize=(18, 5))
# LotArea vs SalePrice
plt.subplot(1, 3, 1)
plt.scatter(ames_subset['LotArea'], ames_subset['SalePrice'], alpha=0.5)
plt.xlabel('LotArea')
plt.ylabel('SalePrice')
plt.title('LotArea vs SalePrice')
# 1stFlrSF vs SalePrice
plt.subplot(1, 3, 2)
plt.scatter(ames_subset['1stFlrSF'], ames_subset['SalePrice'], alpha=0.5)
plt.xlabel('1stFlrSF')
plt.ylabel('SalePrice')
plt.title('1stFlrSF vs SalePrice')
# GrLivArea vs SalePrice
plt.subplot(1, 3, 3)
plt.scatter(ames_subset['GrLivArea'], ames_subset['SalePrice'], alpha=0.5)
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.title('GrLivArea vs SalePrice')

# Show the plots
plt.tight_layout()
plt.show()
````

# Your written answer here - do these seem like good candidates for linear regression?
LotArea vs SalePrice The points show a weak or unclear linear relationship.

1stFlrSF vs SalePrice There seems to be a moderate positive trend — as 1st floor square footage increases, SalePrice tends to increase.

GrLivArea vs SalePrice This shows the strongest linear pattern among the three.



## Step 2: Build a Simple Linear Regression Model

Set the dependent variable (`y`) to be the `SalePrice`, then choose one of the features shown in the subset above to be the baseline independent variable (`X`).

Build a linear regression using StatsModels, describe the overall model performance, and interpret its coefficients.

````python
# Your code here - define y and baseline X
import statsmodels.api as sm
df = pd.DataFrame(ames_subset)
# target variable
y = df['SalePrice']
# baseline predictor variable
X = df[['GrLivArea']]
X = sm.add_constant(X)
````

````python
# Your code here - import StatsModels, fit baseline model, display results
# Fit linear regression model
model = sm.OLS(y, X).fit()
# Display results
print(model.summary())
````

# Your written answer here - interpret model results
R-squared: 0.502 About 50.2% of the variance in SalePrice is explained by GrLivArea

Intercept (const) = 18,570 When GrLivArea = 0, the model predicts a base price of $18,570. Not realistic practically, but useful for model form.

GrLivArea coefficient = 107.13 For every additional square foot, sale price increases by approximately $107, holding all else constant.

## Step 3: Build a Multiple Linear Regression Model

For this model, use **all of** the features in `ames_subset`.

````python
# Your code here - define X
X = df[['LotArea', '1stFlrSF', 'GrLivArea']]
X = sm.add_constant(X)
````

````python
# Your code here - fit model and display results
# Fit the model
model = sm.OLS(y, X).fit()
# Display the results
print(model.summary())
````

# Your written answer here - interpret model results. Does this model seem better than the previous one?
R-squared: 0.565
Now, 56.5% of the variation in SalePrice is explained by the model—an improvement over the previous R² of 0.502

Intercept: -14,310
Base price when all predictors are zero.

GrLivArea ($80.61/sq ft)/ 1st floor ($60.29/sq ft)
Both are highly significant (p < 0.001), with strong positive impacts on price.

LotArea ($0.28/sq ft)
Marginal effect, with a p-value of 0.051. 

## Step 4: Create Partial Regression Plots for Features

Using your model from Step 3, visualize each of the features using partial regression plots.

````python
# Your code here - create partial regression plots for each predictor
# partial regression plots
fig = sm.graphics.plot_partregress_grid(model)
fig.tight_layout(pad=1.0)
plt.show()
````
